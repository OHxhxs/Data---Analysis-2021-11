# Matplotlib - 다양한 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams["font.family"] = 'Malgun Gothic'
```

### 1.산점도(Scatter plot)

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ans.groupby('dataset').agg(['mean','std'])
```

```python
X,Y = np.random.randn(100), np.random.randn(100)
plt.title('산점도')
plt.scatter(X,Y)
plt.grid()
plt.show()
```

```python
colors = np.random.rand(100)
area = np.pi*(15*np.random.rand(100))**2
plt.scatter(X, Y, s=area, c=colors, alpha=0.5)
plt.title('Bubble chart')
plt.show()
```

### 2. 막대 그래프(Bar chart)

```python
s = pd.Series(np.random.rand(10), index = list('abcdefghij'))
plt.bar(s.index,s.values)
plt.title('Bar chart')
plt.show()
```

```python
s.plot(kind = 'bar')
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s = titanic['class'].value_counts()
plt.bar(s.index,s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10), index = list('abcdefghij'))
plt.barh(s.index,s.values)
plt.title('Bar chart')
plt.show()
```

### 3.히스토그램(Histogram)

```python
s = np.random.randn(200)
plt.hist(s)
plt.title('Histogram')
plt.show()
```

```python
plt.hist(s, bins=20)                      # randn(n) n의 개수가 커질수록 종 모양에 
plt.title('bins=20인 경우')               # 모양에 가까워짐
plt.show()
```

### 4.박스 플롯(Box plot)

```python
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
setosa = iris[iris.species == 'setosa'] 
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length, setosa.petal_width], 
            labels= iris.columns[:-1])         #labels=['sl','sw','pl','pw']
plt.title('Setosa 품종의 각 피쳐', fontsize=15) 
plt.show()
```

### 5. 파이차트(Pie Chart)

```python
ratio = [36, 32, 16, 18]
labels = ['사과', '바나나', '멜론', '포도'] 
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
