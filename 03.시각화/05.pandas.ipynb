# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams["font.family"] = 'Malgun Gothic'
```

```python
np.random.seed(2021)
df = pd.DataFrame(np.random.randn(100,3),
                    index=pd.date_range('1/1/2021',periods = 100),
                    columns = list('ABC')).cumsum()
df.plot()
plt.title('Pandas의 plot method 사용 예')
plt.xlabel('일자')
plt.ylabel('Data')
plt.show()
```

# 2. Bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar',rot=0)       #kind는 그래프 모양을 바꿔줄수 있음
plt.title('Setosa 품종의 sepal_length')             # rot는 x인덱스 값을 똑바로 만들어줌
plt.show()                                          # 궁금하면 rot = 90 해볼 것.
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name ='Feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 Feature별 평균')
plt.ylim(0,9)
plt.show()
```

### 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize=(10,10))
plt.title('Setosa 품종의 각 Feature별 Box plot')
plt.show()
```

```python
iris.boxplot(by='species',figsize=(8,10))
plt.suptitle('Iris 품종별 box plot',fontsize=20)
plt.show()
```
