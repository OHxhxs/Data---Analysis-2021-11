# 데이터 프레임 - 그룹분석

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
# 각 품종별 4가지 피쳐의 평균
iris.groupby('species').mean()
```

```python
# 각 품종별 sepal_length의 표준편차

iris[['sepal_length','species']].groupby('species').std()
```

```python
iris.groupby('species')['sepal_length'].std()
```

```python
# 각 품종별 sepla_length의 평균, 표준편차,최댓값, 최솟값
iris.groupby('species')['sepal_length'].agg(['mean','std','max','min'])
```

### tips data 사례

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_percent'] = ((tips.tip/tips.total_bill)*100).round(2)
tips.head()
```

```python
tips.info()
```

```python
tips.describe()      
```

```python
# 성별 데이터 갯수
tips.groupby('sex').count()
```

```python
tips.groupby('sex').size()
```

```python
#성별, 흡연유무별 데이터 갯수
tips.groupby(['sex','smoker']).size()
```

```python
tips.groupby(['sex','smoker']).size()['Male']['Yes']   #['Male','Yes']도 가능.
```

```python
# 성별 tip의 비율의 평균, 최소값, 최대값
tips.groupby('sex')['tip_percent'].agg(['mean','min','max'])
```

```python
# 성별, 흡연유무별 팁 비율의 평균
tips.groupby(['sex','smoker'])['tip_percent'].mean()
```

```python
# 모든 고려요소별 팁 비율의 평균(size만 제외)
tips.groupby(['sex','smoker','time'])[['tip_percent']].mean()
```
